# Q2
Read `LineDetection.jpg` (the image below)




LineDetection.jpg

Use Hough to detect lines.

Note that you should consider the `Threshold`!

Here the threshold is `250`


```python
#TODO
im = cv2.imread('LineDetection.jpg')
im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
plt.imshow(im)
plt.title('Image Name : LineDetection')
plt.axis('off')
plt.show()
```

```python
#TODO
grayim = cv2.cvtColor(im, cv2.IMREAD_GRAYSCALE)

dst = cv2.Canny(grayim, 30, 300)

lines = cv2.HoughLines(dst, 1, np.pi/180, 250)


im2 = im.copy()
for i in lines:
    for ro, theta in i:
        x1 = int(ro * np.cos(theta) + 1000 * (-np.sin(theta)))
        y1 = int(ro * np.sin(theta) + 1000 * (np.cos(theta)))
        x2 = int(ro * np.cos(theta) - 1000 * (-np.sin(theta)))
        y2 = int(np.sin(theta) * ro - 1000 * (np.cos(theta)))
        cv2.line(im2, (x1,y1), (x2,y2), (100, 0, 150), 1)
        
plt.imshow(im2)
plt.title('Image After Line Detection By Hough')
plt.axis('off')
plt.show()
```

```python
def my_hough(dst, treshhold, rho, theta):
    xs, ys = dst.shape
    z = np.sqrt(xs**2 + ys**2)
    max_tresh = int(z)
    
    thetas = np.deg2rad(np.arange(-90, 90, theta))
    rhos = np.linspace(-max_tresh, max_tresh, max_tresh * 2)    
    
    
    hough = np.zeros((max_tresh * 2, len(thetas)))
    points = np.argwhere(dst)              
    
    ro = 0
    for y, x in points:
        for thta in range(len(thetas)):
            ro = x * np.cos(thetas[thta]) + y * np.sin(thetas[thta]) + max_tresh
            hough[int(ro), thta] += 1
    
    less_than_treshold = np.argwhere(hough >= treshhold)
    return [[rhos[x], thetas[y]] for x, y in less_than_treshold]
```

```python
lines = my_hough(dst, 250, 1, 1)
```

```python
im4 = im.copy()
for ro, theta in lines:
    x1 = int(ro * np.cos(theta) + 1000 * (-np.sin(theta)))
    y1 = int(ro * np.sin(theta) + 1000 * (np.cos(theta)))
    x2 = int(ro * np.cos(theta) - 1000 * (-np.sin(theta)))
    y2 = int(np.sin(theta) * ro - 1000 * (np.cos(theta)))
    cv2.line(im4, (x1,y1), (x2,y2), (100, 0, 150), 1)
        
plt.imshow(im4)
plt.title('Image After Line Detection By My Hough')
plt.axis('off')
plt.show()
```

In the cells below you should define two functions that one of them covert RGB to CMYK and another one Convert them vice versa. 

```python
def rgb_to_cmyk(r, g, b, RGB_SCALE = 255, CMYK_SCALE = 100):

  #TODO
    if [r, g, b] == (0, 0, 0):
        return 0, 0, 0, CMYK_SCALE
    r /= RGB_SCALE
    g /= RGB_SCALE
    b /= RGB_SCALE
    
    max_scale = max(r, g, b)
    k = (1 - max_scale)
    
    return round(CMYK_SCALE * (1 - k - r) / (1 - k)), round(CMYK_SCALE * (1 - k - g) / (1 - k)), round(CMYK_SCALE * (1 - k - b) / (1 - k)), round(CMYK_SCALE * k)
```

```python
rgb_to_cmyk(25, 56, 25)
```

Expected Output: (55, 0, 55, 78)

```python
def cmyk_to_rgb(c, m, y, k, CMYK_SCALE = 100, RGB_SCALE = 255):

  #TODO

    c /= CMYK_SCALE
    m /= CMYK_SCALE
    y /= CMYK_SCALE
    k /= CMYK_SCALE

    return round(RGB_SCALE * (1 - c) * (1 - k)), round(RGB_SCALE * (1 - m) * (1 - k)), round(RGB_SCALE * (1 - y) * (1 - k))
```

# Q5
Compute the HSI parametes and V, L and Y according to lectures.

```python
#TODO 
RGB_SCALE = 255
R, G, B = 150, 65, 200
# H = ...
num = (R-G) + (R-B)
denom = 2 * (((R-G)**2 + (R-B) * (G-B))**0.5)
theta = 180 *  np.arccos(num / denom)  / np.pi
if B > G:
    H = 360 - theta
else:
    H = theta

# S = ...
S = 1 - (3 * min(R, G, B)) / (R + G + B), 3

# I = ...
I = (R + G + B) / (3 * RGB_SCALE), 3

# V = ...
max_RGB = max(R, G, B)
min_RGB = min(R, G, B)
V = max_RGB / RGB_SCALE, 3

# L
L = (max_RGB + min_RGB) / (2 * RGB_SCALE)

# Y
Y = 0.299 * R + 0.587 * G + 0.114 * B

print(f'H = {H}')
print(f'S = {S}')
print(f'I = {I}')
print(f'HSI: {H, S, I}')
print(f'V = {V}')
print(f'L = {L}')
print(f'Y = {Y}')
```
