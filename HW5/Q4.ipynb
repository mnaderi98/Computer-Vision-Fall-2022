```python
%matplotlib inline
import cv2
import numpy as np
import math
import matplotlib.pyplot as plt
```

# Q4

```python
#نمایش تصویر
im = cv2.imread('images/img_02.jpg', 0 )
plt.imshow(im, cmap='gray')
plt.show()

s = im.shape
m = s[0]
n = s[1]

#پیدا کردن نقاط غیر سفید
points = [(y, x) for x in range(n) for y in range(m) if im[y, x] != 255]
line = []
for i in range(n):
    for j in range(m):
        if im[j, i] != 255:
            line.append((j, i))

Y = []
X = []
for o in line:
    X.append(o[0])
    Y.append(o[1])

Y_m = np.mean(Y)
X_m = np.mean(X)

s1 = np.multiply(Y, X)
XY_m = np.mean(s1)

X_m_2 = X_m ** 2

s2 = np.power(X, 2)
X_2_m = np.mean(s2)

m = np.float64((Y_m * X_m - XY_m) / (X_m_2 - X_2_m))

c = Y_m - m * X_m

print('y = (' , m , ')x + (' , c, ')' )

x = np.linspace(0, m)

y = m * x + c

plt.plot(x, y, '-r', label=f'y=({m})x+({c})', linewidth=1)

plt.title('line')

plt.xlabel('x')

plt.ylabel('y')

plt.legend(loc='upper left')

plt.imshow(im)

plt.show()
```
