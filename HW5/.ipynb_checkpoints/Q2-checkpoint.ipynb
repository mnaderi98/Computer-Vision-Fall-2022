# Setup

```python
%matplotlib inline
import cv2
import numpy as np
import math
import matplotlib.pyplot as plt
```

# Q2

```python
m = 32
n = 32
im = np.zeros((m, n))
#مختصات نقطه ی سفید که با سعی و خطا بدست آمد
im[31 , 3] = 255
plt.imshow(im, cmap = "gray")
plt.show()
```

```python
image = np.fft.fft2(im)
shiftedimage = np.fft.fftshift(image)


plt.imshow(np.real(shiftedimage), cmap='gray')
plt.show()


plt.imshow(np.imag(shiftedimage), cmap='gray')
plt.show()
```
