```python
# Import required packages
import cv2
import pytesseract
from google.colab.patches import cv2_imshow
from PIL import Image
import matplotlib.pyplot as plt

##################### your code ################################
img = cv2.imread("royan.jpg")
plt.imshow(img, 'gray', vmin = 0, vmax = 255)
plt.title("Original Image")
plt.axis("off")
plt.show()
#Otsu
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
ret,thresh =  cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
otsu_result = thresh
plt.imshow(otsu_result, 'gray', vmin = 0, vmax = 255)
plt.title('Otsu')
plt.axis("off")
plt.show()

#morfology On Otsu:
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 3))
print(kernel)
dilation = cv2.dilate(otsu_result, kernel, 6)
plt.imshow(dilation, 'gray', vmin = 0, vmax = 255)
plt.title("Otsu threshold result dilation")
plt.axis("off")
plt.show()

#adaptive threshold:
thresh =  cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 21, 2)
adaptive_threshold_result = thresh
plt.imshow(adaptive_threshold_result, 'gray', vmin = 0, vmax = 255)
plt.title('adaptive Threshold')
plt.axis("off")
plt.show()

#morfology On adaptive threshold:
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 3))
print(kernel)
dilation_adaptive = cv2.dilate(adaptive_threshold_result, kernel, 6)
plt.imshow(dilation_adaptive, 'gray', vmin = 0, vmax = 255)
plt.title("adaptive threshold result dilation")
plt.axis("off")
plt.show()
################################################################
# Finding contours
## result = result otso|adaptive algorithm
```

```python
print("text on original image")
print(pytesseract.image_to_string(img , lang='fas'))
print("text on Otsu result image")
print(pytesseract.image_to_string(dilation , lang='fas'))
```

```python
print("text on adaptive result image")
print(pytesseract.image_to_string(dilation_adaptive , lang='fas'))
```

```python
# Import required packages
import cv2
import pytesseract
from google.colab.patches import cv2_imshow
from PIL import Image
##################### your code ################################
img = cv2.imread("mypic.png")
plt.imshow(img, 'gray', vmin = 0, vmax = 255)
plt.title("Original Image")
plt.axis("off")
plt.show()
#Otsu
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
ret,thresh =  cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
otsu_result = thresh
plt.imshow(otsu_result, 'gray', vmin = 0, vmax = 255)
plt.title('Otsu')
plt.axis("off")
plt.show()

#morfology On Otsu:
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (2, 1))
print(kernel)
dilation = cv2.dilate(otsu_result, kernel, 100)
plt.imshow(dilation, 'gray', vmin = 0, vmax = 255)
plt.title("Otsu threshold result dilation")
plt.axis("off")
plt.show()

#adaptive threshold:
thresh =  cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 125 , 2)
adaptive_threshold_result = thresh
plt.imshow(adaptive_threshold_result, 'gray', vmin = 0, vmax = 255)
plt.title('adaptive Threshold')
plt.axis("off")
plt.show()

#morfology On adaptive threshold:
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (2, 1))
print(kernel)
dilation_adaptive = cv2.dilate(adaptive_threshold_result, kernel, 3)
plt.imshow(dilation_adaptive, 'gray', vmin = 0, vmax = 1)
plt.title("adaptive threshold result dilation")
plt.axis("off")
plt.show()



################################################################
# Finding contours

```

```python
print("text on original image")
print(pytesseract.image_to_string(img , lang='fas'))

print("text on Otsu result image")
im3 = img.copy()
contours, hierarchy = cv2.findContours(dilation, cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)

for cnt in contours:
  x, y, w, h = cv2.boundingRect(cnt)
	# Drawing a rectangle on copied image
  rect = cv2.rectangle(dilation, (x, y), (x + w, y + h), (0, 255, 0), 2)
	# Cropping the text block for giving input to OCR
  cropped = dilation[y:y + h, x:x + w]
  print(pytesseract.image_to_string(cropped, lang='fas'))

print("text on adaptive result image")

contours, hierarchy = cv2.findContours(dilation_adaptive, cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)
for cnt in contours:
  x, y, w, h = cv2.boundingRect(cnt)
	# Drawing a rectangle on copied image
  rect = cv2.rectangle(dilation_adaptive, (x, y), (x + w, y + h), (0, 255, 0), 2)
	# Cropping the text block for giving input to OCR
  cropped = dilation_adaptive[y:y + h, x:x + w]
  print(pytesseract.image_to_string(cropped, lang='fas'))
```
